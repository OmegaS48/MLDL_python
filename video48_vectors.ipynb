```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set() #gives grids/themes to the plots
```

```python
plt.quiver(0,0,1,1) #(xstart,ystart,anglewithx,anglewithy)
plt.show()
```

```python
plt.quiver(0,0,45,80,scale_units='xy',angles='xy',scale=1,color='#FFDE21')
plt.xlim(-100,100)
plt.ylim(-100,100)
plt.show()
```

```python
plt.quiver(0,0,45,80,scale_units='xy',angles='xy',scale=1,color='#FFDE21')
plt.quiver(0,0,-45,-40,scale_units='xy',angles='xy',scale=1,color='#FF34da')
plt.xlim(-100,100)
plt.ylim(-100,100)
plt.show()
```
