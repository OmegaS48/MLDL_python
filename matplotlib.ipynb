```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
L=["A","B","C","D","E"]
N=[10,20,42,36,45]
ax.bar(L,N)
plt.xlabel("L")
plt.ylabel("N")
plt.show()
```

pie chart


```python
f=plt.figure()
a=f.add_axes([0,0,1,1])
L=["A","B","C","D","E"]
N=[10,20,42,36,45]
a.pie(N,labels=L,autopct="%1.1f%%")
plt.xlabel("L")
plt.ylabel("N")
plt.show()
```
