```python
f=1
n=int(input("Enter number : "))
if(n==0):
  print(f)
else:
  for i in range(1,n+1):
    f=f*i
  print(f)
```

```python
def fact(n):
  f=1
  if(n==0):
    return f
  else:
    for i in range(1,n+1):
      f=f*i
    return f
fact(5)
```
