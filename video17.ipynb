```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
```

```python
y=np.sin(x)
```

```python
plt.show()
```

```python
print(x)
```

```python
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
z=np.cos(x)
plt.plot(x,z)
plt.show()
```

```python
plt.ylabel("yline")
plt.xlabel("xline")
plt.plot(x,z)
plt.show()
```

```python
plt.ylabel("cosine")
plt.xlabel("values")
plt.plot(x,z)
plt.show()
```

```python
k=x**2
```

```python
plt.plot(x,k)
plt.show()
```

```python
p=np.linspace(-20,20,1000)
k=p**3
plt.plot(p,k)
plt.xlabel("Values")
plt.ylabel("Cubic")
plt.show()
```

```python
plt.plot(p,k,'b*')
plt.show()
```

```python
plt.plot(p,np.sin(p),'r+')
plt.plot(p,np.cos(p),'b-')
plt.show()
```

```python
lang=['hindi','marathi','urdu','telugu','english']
people=[5,10,20,40,1]
plt.plot(lang,people)
plt.show()
```

```python
lang=['hindi','marathi','urdu','telugu','english']
people=[5,10,20,40,1]
ax.bar(lang,people)
plt.xlabel('number of people')
plt.ylabel('languages')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
lang=['hindi','marathi','urdu','telugu','english']
people=[5,10,20,40,1]
ax.bar(lang,people)
plt.xlabel('number of people')
plt.ylabel('languages')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
lang=['hindi','marathi','urdu','telugu','english']
people=[5,10,20,40,1]
ax.pie(people,labels=lang,autopct='%1.1f%%')
plt.xlabel('number of people')
plt.ylabel('languages')
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=(1-x**2)**0.5
z=-(1-x**2)**0.5
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(1000)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Reds")
plt.show()
```
